```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bogdanov 2016 - 47Tuc X5

```python
x, y = np.loadtxt('../Bogdanov2016/qLMXB_47TucX5-qLMXB_2016-massradius-hydrogen-1_OLD.txt', unpack=True, comments='#')
plt.plot(x,y)
```

```python
# READ 1SIGMA AND 2SIGMA COUNTOURS 
r1, m1, r2, m2 = np.loadtxt('../Bogdanov2016/qLMXB_47TucX5-qLMXB_2016-massradius-hydrogen-1.txt', unpack=True, comments='#')

# Remove NaN from file
r1 = r1[~np.isnan(r1)]
m1 = m1[~np.isnan(m1)]
r2 = r2[~np.isnan(r2)]
m2 = m2[~np.isnan(m2)]

if len(r1)!=len(m1):
    pass
else:
    plt.plot(r1,m1, label='1 sigma')
    
if len(r2)!=len(m2):
    pass
else:
    plt.plot(r2,m2, label='2 sigma')
plt.legend()
```
